## Q5

## Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap

```python
data = sns.load_dataset("tips")

a = print(data.corr(numeric_only = True))
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate a synthetic dataset with correlated features
n_samples = 500
mean = [0, 0, 0, 0, 0]
cov = [[1, 0.8, 0.6, 0.4, 0.2],    # Correlation matrix
       [0.8, 1, 0.75, 0.55, 0.35],
       [0.6, 0.75, 1, 0.7, 0.45],
       [0.4, 0.55, 0.7, 1, 0.65],
       [0.2, 0.35, 0.45, 0.65, 1]]

# Generate random data based on the covariance matrix
data = np.random.multivariate_normal(mean, cov, size=n_samples)

# Create a pandas DataFrame
columns = ['Feature_1', 'Feature_2', 'Feature_3', 'Feature_4', 'Feature_5']
df = pd.DataFrame(data, columns=columns)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

## Q2

## Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Viridis256


output_notebook()


np.random.seed(42)
x = np.random.rand(100) * 100
y = np.random.rand(100) * 100
sizes = np.random.randint(10, 40, 100)
colors = np.random.rand(100) * 100


source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))


p = figure(title="Bokeh Scatter Plot with Sizes and Colors",
           x_axis_label='X values', y_axis_label='Y values',
           width=700, height=400)


mapper = linear_cmap(field_name='colors', palette=Viridis256, low=min(colors), high=max(colors))


p.scatter('x', 'y', size='sizes', color=mapper, source=source, legend_label="Color based on 'colors' column", fill_alpha=0.6)


p.legend.location = "top_left"
p.grid.grid_line_alpha = 0.3

show(p)
```

## Q3

## Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource


output_notebook()


fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]


source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))


p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label="Fruits", y_axis_label="Count",
          )


p.vbar(x='fruits', top='counts', width=0.7, source=source,
       legend_field='fruits', line_color='white',
       fill_color=factor_cmap('fruits', palette=['#ff9999', '#ffcc99', '#99ccff', '#99ff99'], factors=fruits))

p.legend.orientation = "horizontal"
p.legend.location = "top_center"


show(p)
```

## Q4

## Create a Bokeh histogram to visualize the distribution of the given data.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

source = ColumnDataSource(data=dict(
    edges=edges[:-1],
    top=hist
))

output_file("histogram.html")

p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency',
           tools="", background_fill_color="#fafafa")
p.quad(top='top', bottom=0, left='edges', right='edges', source=source,
       fill_color="skyblue", line_color="white", alpha=0.7)
show(p)
```

##  Create a Bokeh heatmap using the provided dataset.

```python
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
```

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
from bokeh.io import output_notebook
from bokeh.layouts import column


data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)


x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()


source = ColumnDataSource(data=dict(
    x=x_flat,
    y=y_flat,
    z=z_flat
))
output_file("heatmap.html")

p = figure(title="Heatmap of Random Data", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1),
           tooltips=[('X', '@x'), ('Y', '@y'), ('Value', '@z')],
           width=600, height=600)

mapper = linear_cmap(field_name='z', palette=Viridis256, low=min(z_flat), high=max(z_flat))

p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
       line_color=None, fill_color=mapper)

color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

show(p)
```
